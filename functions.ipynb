```python
# default args
def cleanText(data, to_remove = "@$#%"):
    newdata = data                  #make a copy of data
    for char in to_remove:          #loop of each char to remove
        newdata = newdata.replace(char, '') #use replace to remove char
    return newdata                  #return newdata

cText = cleanText("He#ll#o@")       #default chars used for removing
print(cText)
cText2 = cleanText("H@!()#!)@i", to_remove = '@!#()')   #updated chars
print(cText2)
```

### Variable Arguments in python

A function can have variable arguments in python using * symbol before a parameter in the function defination. This allows the function to accept any positional arguments as a tuple.

```python
# A function that can multiply n parameters
def multiplier(*numbers):
    result = 1
    for value in numbers:
        result *= value
    return result

print(multiplier())         #no value
print(multiplier(2, 3, 4))  #with few arguments
print(multiplier(2, 3, 4, 3, 5 ,6, 1, 2, 2, 5, 56, 2))  #with many arguments
```

```python
def operate(*args, operator="*"):
    if operator == '*' :
        result = 1
        for num in args:
            result *= num
        return result
    elif operator == '*' :
        result = 0
        for num in args:
            result+=num
        return result
    else:
        return "Wrong operator give, please provide + or *"
    
print(operate(1, 2, 3, 4))
print(operate(1, 2, 3, 4, operator = '*'))
print(operate(1, 2, 3, 4, operator = '-'))

```

### Generators or lazy functions
special type of function that will not evaluate and return all values at once, instead it will return onw value ata a time and only when requested

```python
for i in range(1, 6): print(i)

print(list(range(1, 10)))
```

```python
# creating a generator (use yield instead of return)
def squared(data):
    for i in data:
        yield i**2          #yield store all result in memory to be
                            #used later and dont retuen a single value

#squared(data=[1,2,3,4,5,6,7,8])     #this will not give you result

#this is how we use generator (getting 1 value at a time)
for num in squared([2, 3, 3, 5, 6, 7]):
    print("Square", num)

# or
#or you can directly convert the output of generator to a data structures
print(list(squared([22, 33, 55])))  #this calculates and converts all value to 
```

```python
def getAcronyms(data = []):
    for name in data:
        acronym = "".join(name.split())
        yield acronym

names = []
list(getAcronyms[names])
```
